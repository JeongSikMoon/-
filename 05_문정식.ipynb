```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
# 크롬 드라이버에 옵션 주기
options = webdriver.ChromeOptions()
options.add_argument('--headless') # 인터페이스 없는
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--window-size=1920x1080')
```

## 2. html을 받아오기


```python
# https://datalab.naver.com/keyword/realtimeList.naver

# driver 변수를 Chrome webdriver 클래스로 만든다. 
driver = webdriver.Chrome('C:/Users/JY-JS/Downloads/chromedriver_win32/chromedriver',options=options)

# 3초간 기다리자.
driver.implicitly_wait(3)

# 페이지에서 소스 가져오기
driver.get('https://datalab.naver.com/keyword/realtimeList.naver')

# 5초간 기다려서 javascript 내용이 뜰때까지 기다리기
time.sleep(5)

# 페이지 소스를 html이라는 변수에 저장하기
html = driver.page_source
```
